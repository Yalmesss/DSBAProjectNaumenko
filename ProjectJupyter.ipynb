```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("final_cars_datasets.csv" )
del df['Unnamed: 0']
```

```python
df
```

```python
print('Information about prices of the cars:', "\n")
print(' 1.The mean value of the car price is' , round(df['price'].mean(), 2), 'hundreds of dollars','\n','2.The median value of the car price is', round(df['price'].median(), 2), 'hundreds of dollars','\n','3.The value of the standard deviation from the average car price is',round(df['price'].std(), 2), 'hundreds of dollars')

sns.set_style("darkgrid")
ax= sns.violinplot(y=df['price'],inner= 'quartile', palette="flare", as_cmap=True)
```

```python
print('Information about miles traveled by vehicles:', "\n")
print(' 1.Average mileage of cars is', round(df['mileage'].mean(),2),'\n','2.The median value of mileage among cars is',round(df['mileage'].median(), 2),'\n','3.The value of the standard deviation from the average car mileage is',round(df['mileage'].std(), 2))

box= plt.boxplot(df['mileage'], patch_artist=True, sym='.')
plt.style.use('classic')
plt.ylabel('Miles')
plt.title('Mileage')
plt.setp(box["boxes"], facecolor='pink')
plt.show()
```

```python
print('Information about the total volume of the cylinders in the engine:', '\n')
print(' 1.The mean value the of the volume of the cylinders in the engine is',round(round(df['engine_capacity'].mean(), 1)/1000,1),'litres','\n','2.The median value of the volume of the cylinders in the engine among cars is',round(round(df['engine_capacity'].median(),1)/1000,1),'litres','\n', '3.The value of the standard deviation from the average volume of the cylinders in the engine of the car is',round(round(df['engine_capacity'].std(), 2)/1000,2),'litres')

plt.style.use('classic')
plt.hist(round(df['engine_capacity']/1000,2), color='purple')
plt.xlabel('Engine capacity')
plt.ylabel('Count')
plt.title('Engine capacity in litres')
plt.xlim(left=0, right=6)
plt.show()
```

```python
sns.boxenplot(x=df['drive'], y=df['price'],
              palette=["m", "g"], 
              scale="linear", );
plt.title('Distribution by the type of drive and  price.''\n');
```

```python
print(' From the graph it becomes clear that cars with two driving wheels have a much wider price range than cars with other types of drives.','\n','At the same time, the median price for cars with all-wheel drive is higher than for cars with other types of drives. ')
```

```python
sns.boxplot(x=df['fuel'], y=df['price'],
            hue=df['hand_drive']);
plt.title('Distribution by the type of fuel, price and the hand of drive.''\n');
```

```python
print(' Speaking of cars driving on gasoline, you can notice that RHD is much greater variation in price, which is likely due to their large number, the median price is $ 1,000, while the median price of the machine with the left rudder driving on gasoline a few more(about 50 hundred dollars).','\n','Cars with steering wheel on the center use only diesel fuel. The difference in the median price of cars with right and left hand drive driving on diesel fuel is very different (about 400 hundreds of dollars).','\n','Cars driving on other types of fuel are represented only with right hand drive')
```

```python

fig= sns.relplot(x=df["year"]-2000, y=df['price'], hue=df['mark'], 
            alpha=.5,
            height=6);
plt.title('Distribution by the mark,year and price.''\n');
```
